```python
## removendo as colunas desnecessarias
df =  df.drop(columns= ['Foto do trabalho administrativo', 'Qual o PDV será auditado?', 'Durante meu trabalho em loja existia mais alguém presente?', 'Função do colaborador:', 'Qual o nome do colaborador?', 'Qual cliente estamos auditando nesse momento?', 'Foto do antes:', 'Foto do depois:', 'Para quem foi reportada?','Foto do antes:.1','Foto do depois:.1','Foto do Status geral (sobre o produto/familia ou pontos positivos que desejo mostrar)','Comentários gerais sobre o cliente:', 'Para qual cliente estou trabalhando?', 'Quem é o promotor?'])
```

```python
## verificando duplicadas
df.duplicated().sum()
```

```python
## separando a planilha em duas tabelas, administrativo e operacional, conforme excel do emerson
df_adm = df[df['O trabalho que vou prestar hoje é administrativo?'] == 'SIM']
```

```python
df_ope = df[df['O trabalho que vou prestar hoje é administrativo?'] == 'NÃO']
```

```python
df_adm.head()
```

```python
## removendo as colunas desnecessarias
df_adm =  df_adm.drop(columns= ['Qual o tipo de oportunidade encontramos nesse momento?', 'Durante o meu trabalho em loja foi identificada alguma melhoria para este cliente?', 'Qual o tipo de melhoria?', 'Esta melhoria foi:'])
```

```python
df_ope =  df_ope.drop(columns= ['O trabalho que vou prestar hoje é administrativo?', 'Qual trabalho administrativo estarei fazendo?', 'Comentários (a respeito do trabalho adm. realizado)'])
```

```python
## preenchendo os valores ausentes 
df_adm.fillna('', inplace=True)
df_ope.fillna('', inplace=True)
```
